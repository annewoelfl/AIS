```python
import os

def delete_files_in_second_folder(folder1, folder2):
    # Get the list of files in folder1 and folder2
    files_in_folder1 = set(f for f in os.listdir(folder1) if f.endswith('.csv'))
    files_in_folder2 = os.listdir(folder2)

    # Loop through files in folder2
    for file in files_in_folder2:
        if file.endswith('.csv') and file in files_in_folder1:
            file_path = os.path.join(folder2, file)
            try:
                os.remove(file_path)  # Delete the file
                print(f"Deleted: {file}")
            except Exception as e:
                print(f"Error deleting {file}: {e}")

# Example usage:
folder1 = './folders_to_compare/6 days set'
folder2 = './folders_to_compare/9 days set'

delete_files_in_second_folder(folder1, folder2)
```

```python
import os

# Specify the path to your folder
folder_path = r'./folders_to_compare/12 days set'

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)

    # Check if it's a file and the filename length is greater than 19 characters
    if os.path.isfile(file_path) and len(filename) > 19:
        new_filename = filename[19:]  # Remove the first 19 characters
        new_file_path = os.path.join(folder_path, new_filename)

        # If the new filename already exists, append '_1' to the filename
        counter = 1
        while os.path.exists(new_file_path):
            # Create a new filename by appending _1, _2, etc.
            base_name, ext = os.path.splitext(new_filename)
            new_filename = f"{base_name}_{counter}{ext}"
            new_file_path = os.path.join(folder_path, new_filename)
            counter += 1

        # Rename the file
        os.rename(file_path, new_file_path)

        print(f'Renamed: {filename} -> {new_filename}')
```
